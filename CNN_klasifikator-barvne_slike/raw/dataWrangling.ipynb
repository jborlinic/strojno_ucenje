```python
import pandas as pd

data = pd.DataFrame()

labels = pd.DataFrame()

for i in range(5):
    readData = pd.read_pickle('data/cifar-10-batches-py/data_batch_%d' %(i + 1))
    
    tempDFdata = pd.DataFrame(readData['data'])
    data = data.append(tempDFdata, ignore_index = True)
    
    tempDFlabels = pd.DataFrame(readData['labels'])
    labels = labels.append(tempDFlabels, ignore_index = True)
    
data.to_csv('data/train_images.csv', index=False, header=False)
labels.to_csv('data/train_labels.csv', index=False, header=False)
```

```python
test = pd.read_pickle('data/cifar-10-batches-py/test_batch')
test_data = pd.DataFrame(test['data'])
test_labels = pd.DataFrame(test['labels'])

test_data.to_csv('data/test_images.csv', index=False, header=False)
test_labels.to_csv('data/test_labels.csv', index=False, header=False)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
for i in range(20):
    image = np.array(test_data.iloc[i])
    image = np.reshape(image, (3,1024))
    image = np.transpose(image)
    image = np.reshape(image, (32,32,3))
    imgplot = plt.imshow(image)
    plt.show()
```
