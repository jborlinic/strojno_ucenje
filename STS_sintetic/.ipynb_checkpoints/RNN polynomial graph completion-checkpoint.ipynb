# Recurrent neural network for the completion of sequential polinomial graph data

This is the third synthetic test case for RNN (recurrent neural network) model structure. It is a stepping stone towards a sequence to sequence RNN to predict (complete) GPS tracks.
This test case builds upon a simpler RNN model to complete polynomial graph data, this time we allow the sequence of graph points in both directions $x_n > x_{n-1}> ... > x_0$ and $x_0 > x_1 > ... > x_n$.

First two synthetic models (_notebooks in Slovene_):
- [Completion of quadratic functions](https://github.com/jborlinic/machine_learning/blob/master/STS_sintetic/RNN%20za%20dokon%C4%8Danje%20izrisov%20grafov%20kvadratnih%20funkcij.ipynb)  
- [Completion of 1-directional polynomials](https://github.com/jborlinic/machine_learning/blob/master/STS_sintetic/RNN%20za%20dokon%C4%8Danje%20izrisov%20grafov%20polinomov.ipynb)  

As our data set we generate sequences of $(x,y)$ points representing polynomial graphs. These should have a similar structure as our GPS data. This means our neural net will take as an input a sequence of $15$ $(x,y)$ points and return a sequence of the following $5$. 

This is our second synthetic data set for an sequence to sequence (STS) RNN model.

## Generating and data preprocessing 

We generate 100000 random polynomials with the help of the root form of the polynomial function with five random roots: $f = a(x-x_0)\cdot b(x-x_1)\cdot c(x-x_2)\cdot d(x-x_3)\cdot e(x-x_4)$, where $x_0,x_1, x_2, x_3, x_4 \in [10,80]$ random roots and $a,b,c,d,e \in \{-1,1\}$.  

We use this terms to generate the random polynomials.  

Eventually we want to use the model on GPS data so we extract $25$ sequential points of each randomly generated polynomial.  
Later we will fragment these $25$ points to $25 - 20 = 5$ sub-sequences, normalize them to the interval $[0,1]\times[0,1]$ and split them into three data subsets, the training set, validation set and test set.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_polinomial_data(n_to_gen, finess=50, show_graphs=False):
    """
    A method that generates n_to_gen polinomials with finess number of points equaly displaced on 
    the interval (0,100).
    Parameters:
        - n_to_gen (int), number of polinomials to generate
        - finess (int), the number of points for each polinomial
        - show_graph (bool), if the function should also print the generated polinomials
    """
    x = np.arange(finess)
    x = x / (finess / 100)
    data = []
    for _ in range(n_to_gen):
        x0 = np.random.randint(80) + 10
        x1 = np.random.randint(80) + 10
        x2 = np.random.randint(80) + 10
        x3 = np.random.randint(80) + 10
        x4 = np.random.randint(80) + 10
        
        #print(x0, x1, x2, x3, x4)
        p =(np.power(-1, x0) * (x - x0) * 
            np.power(-1, x1) * (x - x1) * 
            np.power(-1, x2) * (x - x2))# * 
            #np.power(-1, x3) * (x - x3) *
            #np.power(-1, x4) * (x - x4))
        
        x = np.reshape(x, (-1,1))
        y = np.reshape(p, (-1,1))
    
        points = np.append(x, y, axis=1)
        
        if show_graphs:
            plt.scatter(points[:,0], points[:,1])
            plt.show()
              
        data.append(points)
        
    return np.array(data)
```

```python
def normalize(data):
    """
    Method that normalizes the generated data.
    Parameters:
        - data, (np.array of shape (n_of_functions, n_of_points, 2))
    """
    for i in range(2):
        maxI = np.amax(data[:,:,i])
        minI = np.amin(data[:,:,i])
        data[:,:,i] = (data[:,:,i] - minI) 
        data[:,:,i] = data[:,:,i] / (maxI - minI)
    
    return data
```

As in the former two test cases, we generate $100000$ random polynomials, where each is represented by $25$ points.  
We add the show_graph=False parameter to omit the plotting of all of these polynomials. However, we do want to check if our methods did their job correctly, therefore we use the next for loop to plot $5$ random polynomials from our data set. 

```python
gen_data = normalize(generate_polinomial_data(100000, finess=25, show_graphs=False))

print(gen_data.shape)
```

```python
for _ in range(5): # spremeni range(5), ce zelis vec ali manj izrisanih grafov
    a = np.random.randint(gen_data.shape[0])
    points = gen_data[a]
    plt.scatter(points[:,0], points[:,1])
    plt.show()
```

If we look at the plotted graphs, we can see our methods did their job and we have a well generated data set.

The next step is to transform each sequence of $25$ points into $5$ two-part sub-sequences of $20$ and $5$ points to represent our input & output data.

```python
def partition_data(data, ts_in, ts_out):
    """
    The method partition the "track" data of shape (n_of_tracks, n_of_points_per_track, n_of_dim_per_point) into 
    two datasets of tracks representing the input and output of our model, with length ts_in and ts_out respectively.
    Parameters:
        - data (np.array), the dataset to be partitioned of shape (n_of_tracks, n_of_points_per_track, n_of_dim_per_point)
        - ts_in (int), length of the needed input dataset
        - ts_out (int), length of the needed output dataset

    NOTE: ts_in + ts_out < n_of_points_per_track
    """
    print(data.shape)
    data_in = []
    data_out = []
    r = data.shape[1]
    for f in data:
        for i in range(r//(ts_in + ts_out)):
            backwards = np.random.randint(2)
            
            if backwards % 2 == 0:
                f = np.flip(f,0)
                
            new_in = f[i:i + ts_in]
            new_out = f[i + ts_in:i + ts_in + ts_out]    
            data_in.append(new_in)
            data_out.append(new_out)
            
    data_in = np.array(data_in)
    data_out = np.array(data_out)
    
    return data_in, data_out
    
data_in, data_out = partition_data(gen_data, 15, 5)
```

Another for loop to check if our split of the data was correct.

```python
for _ in range(5):
    i = np.random.randint(data_in.shape[0])
    print(i)
    plt.scatter(data_in[i,:,0],data_in[i,:,1], color='blue')
    plt.scatter(data_out[i,:,0],data_out[i,:,1], color='red')
    plt.show()
```

The last two steps at our data preparation involve:
- splitting the data set into three disjointed subsets (training, validation, test). These should be split in the ratio ($80%$, $10%$, $10%$) respectively.  
- saving the gathered data into the directory: ../.datasets/sintetic_STS/polinomials/

```python
def split_data(data, train_per=0.8, val_per=0.1, test_per=0.1):
    """
    The method splits our dataset over axis=0 into three datasets, training data, validation data and test data.
    Parameters:
        - data (np.array), the dataset to be split,
        - train_per (float), the % of the data to be used for the training set,
        - val_per (float), the % of the data to be used for the validation set,
        - test_per (float), the % of the data to be used for the test set.
    """
        
    train_ind = int(data.shape[0] * train_per)
    val_ind = int(data.shape[0] * (train_per + val_per))
    train_data = data[:train_ind]
    val_data = data[train_ind:val_ind]
    test_data = data[val_ind:]
    
    return train_data, val_data, test_data
```

## STS model
STS (sequence to sequence) models were designed to work with sequential data. They are a specific set of RNNs, where the models outputs, not only inputs, are sequences of data.  

A very good description can be found at [Microsoft's GitHubu for STS models in CNTK](https://github.com/Microsoft/CNTK/wiki/Sequence-to-Sequence-%E2%80%93-Deep-Recurrent-Neural-Networks-in-CNTK-%E2%80%93-Part-1).  

Here we implemented a SQS model to predict the next $5$ points according to a sequence of $15$ former points. This means it receives a 3-D input array of the shape: (n. of examples, $15$ points, $2$ coordinates).

```python
import keras
from keras.layers import LSTM, Dense, Dropout, TimeDistributed, RepeatVector
```

### Model definition
Let's define a sequential model that consists of an encoder, a few intermediate layers and a decoder.

The encoder consists of a [LSTM](http://colah.github.io/posts/2015-08-Understanding-LSTMs/) layer with *hidden_naurons* ($=2000$) neurons. The intermediate layers are represented by a dense ($2000 \times 2000$) layer with the [ReLu](https://www.mathworks.com/help/nnet/ref/relulayer-class.html?requestedDomain=www.mathworks.com) activation function, a dropout layer with $50%$ keep percentage. Next we have a repeat vektor, which makes $5$ copies of the existing vector and this is the input to our decoder. The decoder consists of another [LSTM](http://colah.github.io/posts/2015-08-Understanding-LSTMs/) layer with *hidden_naurons* ($=2000$) neurons and at the end there is a dense layer in a time distributed wrapper (each value is computed separately) and this is how we produce our sequence of $5% predicted points.

```python
def seqToSeq_model(n_of_features, input_sequence_length, output_sequence_length):
    
    hidden_neurons = 2000
    model = keras.models.Sequential()

    ##Encoder
    model.add(LSTM(hidden_neurons,
              input_shape=(input_sequence_length, n_of_features),
              return_sequences=False
              ))
    print(model.outputs)
    model.add(Dense(hidden_neurons, activation='relu'))
    model.add(Dropout(0.5))

    print(model.outputs)
    model.add(RepeatVector(output_sequence_length))
    print(model.outputs)
    
    ##Decoder
    model.add(LSTM(hidden_neurons,
              return_sequences=True))

    print(model.outputs)
    model.add(TimeDistributed(Dense(n_of_features, activation='relu')))
    print(model.outputs)

    return model
```

When we train our model, we want to have the ability to save the resulting model and load it afterwards. This is why we have the following two methods.


```python
from keras.models import model_from_json

def load_model(model_dir):
    """
    This method loads a previously traned model from model_dir.
    Parameters:
        - model_dir (string), a directory where the model and its weights are saved.
    """
    try:
        f = open('%smodel' %model_dir, 'r')
        json_string = f.read()

        model = model_from_json(json_string)
        model.load_weights('%sweights' %model_dir)
        print('Model loaded!')

        return model

    except FileNotFoundError:
        print('Could not load model!')


def save_model(model_dir, model):
    """
    This method saves the model to the model_dir directory.
    Properties:
        - model (keras.model), a trained keras model,
        - model_dir (string), the directory to save the model to.
    """
    json_string = model.to_json()

    f = open('%smodel' %model_dir, 'w+')
    f.write(json_string)
    f.close()

    model.save_weights('%sweights' %model_dir)
    print("Model saved!")
```

It is finally the time to train our model.  
For this we need to compile the loaded model (set the loss function and our optimizer) and then use the model.fit function to train the model.


```python
MODEL_DIR = 'model/bi_dir_polinomial/'

model = seqToSeq_model(2, 15, 5)

model.compile(loss='mean_squared_error', optimizer='Adam')
model.fit(train_data_in, train_data_out, epochs=3, batch_size=100, validation_data=(val_data_in, val_data_out))
```

The model.predict method uses our trained model to predict the next values of a sequence of numbers. This is done on the validation or test sets since these are the sets our model has not been trained on, so they give us a more accurate representation of the effectiveness of the model. Finally we use the for loop to plot a few ($50$) polynomials with the predictions to see how good our model is (in the plots the blue dots represent the input sequence, the red dots represent the expected output and the green dots are the predictions our model has made.

```python
#save_model(MODEL_DIR, model)
MODEL_DIR = 'model/bi_dir_polinomial/'

model = load_model(MODEL_DIR)
model.compile(loss='mean_squared_error', optimizer='Adam')
model.fit(train_data_in, train_data_out, epochs=1, batch_size=100, validation_data=(val_data_in, val_data_out))

pred = model.predict(test_data_in)
```

```python
for _ in range(50):
    i = np.random.randint(val_data_in.shape[0])
    print(i)
    plt.scatter(test_data_in[i,:,0],test_data_in[i,:,1], color='blue')
    plt.scatter(test_data_out[i,:,0],test_data_out[i,:,1], color='red')
    plt.scatter(pred[i,:,0], pred[i,:,1], color='green')
    plt.show()
```

The plots show us the effectiveness of our model. After only $10$ epochs of training we can see that our model has learned to recognize the correct polynomial and make a very accurate prediction of the last $5$ points.  
The most interesting parts are the predictions of the extremes of certain polynomials which are not a trivial "linear" prediction, so our model has gained deeper insight in the polynomial graphs. Similar results were shown in the second test case as well however, this example confirms the expected need of a significantly deeper network just to gain the ability to learn the direction of the sequence.  

At the end, let's just save the final model with the save_model function.

```python
MODEL_DIR = 'model/bi_dir_polinomial/'
save_model(MODEL_DIR, model)
```
