```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_polinomial_data(n_to_gen, finness):
    n = finness
    x = np.arange(n)
    x = x / (finness / 100)
    for _ in range(n_to_gen):
        x0 = np.random.randint(80) + 10
        x1 = np.random.randint(80) + 10
        x2 = np.random.randint(80) + 10
        x3 = np.random.randint(80) + 10
        x4 = np.random.randint(80) + 10
        
        print(x0, x1, x2, x3, x4)
        p =(np.power(-1, x0) * (x - x0) * 
            np.power(-1, x1) * (x - x1) * 
            np.power(-1, x2) * (x - x2) * 
            np.power(-1, x3) * (x - x3) *
            np.power(-1, x4) * (x - x4))
        plt.scatter(x, p)
        plt.show()

n = 1000000        
generate_polinomial_data(10, n)
```
